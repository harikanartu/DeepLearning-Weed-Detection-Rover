```python
import shutil

# Create train/valid directories for images and labels if they don't exist
os.makedirs('./train/images', exist_ok=True)
os.makedirs('./train/labels', exist_ok=True)
os.makedirs('./valid/images', exist_ok=True)
os.makedirs('./valid/labels', exist_ok=True)
os.makedirs('./test/images', exist_ok=True)
os.makedirs('./test/labels', exist_ok=True)

# Move images and labels from the dataset to the correct structure
for img in os.listdir(os.path.join(train_path, 'images')):
    shutil.move(os.path.join(train_path, 'images', img), './train/images')
    shutil.move(os.path.join(train_path, 'labels', img.replace('.jpg', '.txt')), './train/labels')

for img in os.listdir(os.path.join(valid_path, 'images')):
    shutil.move(os.path.join(valid_path, 'images', img), './valid/images')
    shutil.move(os.path.join(valid_path, 'labels', img.replace('.jpg', '.txt')), './valid/labels')

for img in os.listdir(os.path.join(test_path, 'images')):
    shutil.move(os.path.join(test_path, 'images', img), './test/images')
    shutil.move(os.path.join(test_path, 'labels', img.replace('.jpg', '.txt')), './test/labels')

print("Dataset organized into train, valid, and test directories.")
```

```python
# Write the data.yaml file for YOLOv8
yaml_content = """
train: ./train/images
val: ./valid/images
test: ./test/images
nc: 1  # Number of classes (weed plants)
names: ['weed']  # List of class names
"""

# Save it as data.yaml
with open("data.yaml", "w") as f:
    f.write(yaml_content)

print("data.yaml file created!")
```

```python
# Import YOLOv8 library and train the model
from ultralytics import YOLO

# Initialize YOLOv8 model (use a pre-trained model for faster training)
model = YOLO('yolov8n.pt')  # Choose 'yolov8n.pt', 'yolov8s.pt', etc.

# Train the model
model.train(data='data.yaml', epochs=5, imgsz=640, batch=16, name='weed-detection')

print("Training complete!")
```
